```python
# Write your code for creating the data frame df_by_decade here 
df_by_decade = pd.DataFrame(movies.groupby('decade')['num_voted_users'].sum())
```

```python
# Write your code for plotting number of voted users vs decade

df_by_decade.reset_index(level = 'decade',inplace = True) #Index changed to column for use in bar graph

plt.figure(figsize = (10,8))
plt.bar(df_by_decade['decade'],df_by_decade['num_voted_users'])
plt.xlabel('Decade', fontsize = 12)
plt.ylabel('Voted Users', fontsize = 12)
plt.ticklabel_format(axis = 'y',style = 'plain') #Removes the scientific notation on the y-axis resulting from large numbers
plt.show()
```
